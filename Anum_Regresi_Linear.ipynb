```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("seaborn")
```

```
def linearRegresion(data):
  '''
  indeks[0] -> response variable -> x
  indeks[1] -> predictor variable -> y
  '''
  x2=[]
  y2=[]
  xy=[]
  n = len(data[0])
  for x in data[0]:
    x2.append(x**2)
  for y in data[1]:
    y2.append(y**2)
  i=0;
  while(i<n):
    dump = data[0][i]*data[1][i]
    xy.append(dump)
    i+=1
  jmlhx = sum(data[0])
  jmlhy = sum(data[1])
  jmlhx2 = sum(x2)
  jmlhy2 = sum(y2)
  jmlhxy = sum(xy)
  a = ((jmlhy*jmlhx2)-(jmlhx*jmlhxy))/(n*jmlhx2-(jmlhx**2))
  b = ((n*jmlhxy)-(jmlhx*jmlhy))/(n*jmlhx2-(jmlhx**2))
  return(a,b)
```

```
def gambarGrafik(dataProses):
  a,b = linearRegresion(dataProses)
  print("Nilai a adalah %.4f"%(a))
  print("Nilai b adalah %.4f"%(b))
  def f1(keanggotaan,a,b):
    hit = []
    for x in keanggotaan:
      y = b*x+a
      hit.append(y)
    return(hit)
  plt.scatter(dataProses[0],dataProses[1],color='blue',label='Data Sebenarnya',s=30 )
  plt.plot(dataProses[0],f1(dataProses[0],a,b),'r-o',label='Hasil Regresi',linewidth=2)
  plt.title("Hasil Regresi Linear")
  plt.ylabel(yl)
  plt.xlabel(xl)
  plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
  fig = plt.figure(1)
  plt.show()
```

```
Jml_produksi = [30.0, 20.0, 60.0, 80.0, 40.0, 50.0, 60.0, 30.0, 70.0, 60.0]
Jam_kerja = [73.0, 50.0, 128.0, 170.0, 87.0, 108.0, 135.0, 69.0, 148.0, 132.0]
xl = "Jumlah Produksi"
yl = "Jam Kerja"
gambarGrafik([Jml_produksi,Jam_kerja])
```
