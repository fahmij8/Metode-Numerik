```
#@title Import module
import math 
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integral

from math import *
from matplotlib import style
style.use("seaborn")
```

```
#@title Midpoint Rule
# Midpoint's Rule
def midpoint(f, a, b, n):
 h = float(b-a)/n
 #Remove comment dibawah untuk mengetahui perhitungan komputer
 #print(h)
 result =0
 for i in range(n):
  result += f((a + h/2.0) + i*h)
  result *= h
  #Remove comment dibawah untuk mengetahui perhitungan komputer
  #print(result)
 return result

def realIntegration(f,a,b):
  result = integral.quad(f, a, b)
  return result
```

```
#@title Fungsi tampilan grafik
#Area integral, dengan matplotlib
import warnings
warnings.filterwarnings("ignore")

def integArea(eq, upperLim, bottomLim, breakUp):
    f = eq
    a = upperLim; b = bottomLim; N = breakUp
    x = np.linspace(a,b,N+1)
    X = np.linspace(a,b,N*N+1)
    do = np.linspace(a,b,1000)
    Y = f(X)


    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.plot(do, f(do), color="#0000ff")
    plt.title('Grafik fungsi f(x)')
    plt.ylim(bottom=0)

    plt.subplot(1,3,2)
    plt.plot(do, f(do), color="#0000ff")
    plt.title('Pendekatan Integral fungsi f(x)')
    plt.fill_between(do, f(do), where=[(do >= a) and (do <= b) for do in do], color="#0f0f0f50")
    plt.ylim(bottom=0)

    plt.subplot(1,3,3)
    plt.plot(X,Y,'b')
    x_mid = (x[:-1] + x[1:])/2 # Midpoints
    y_mid = f(x_mid)
    plt.plot(x_mid,y_mid,'r.',markersize=10)
    plt.bar(x_mid,y_mid,width=(b-a)/N,color="#0f0f0f50")
    plt.title('Midpoint Rule, N = {}'.format(N))
```

## Soal No.1
Gunakan aturan trapezoidal dari **n=3** dan fungsi

$f(x) = \int_{-1}^{2}\left( x^{2}+1 \right) \mathrm{d}x $


```
#@title Jawaban
def f(x):
  return x**2+1

a = -1
b = 2
n = 3

integArea(f, a, b, n)
print("Metode midpoint = ", midpoint(f, a, b, n))
print("integral = " + str(realIntegration(f,a,b)[0]) + "\n")
```

## Soal No.2
Gunakan aturan trapezoidal dari **n=3** dan fungsi

$f(x) =  \int_{0}^{\pi} sin(x) \mathrm{d}x $



```
#@title Jawaban
def f(x):
  return np.sin(x)

a = 0
b = np.pi
n = 3

integArea(f,a,b,n)
print("Metode midpoint = ", midpoint(f, a, b, n))
print("integral = " + str(realIntegration(f,a,b)[0]))
```
